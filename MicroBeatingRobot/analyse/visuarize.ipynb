```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import re
%matplotlib tk
```

### Generate DataFrame

```python
path = "/home/tokoro/Desktop/MicroBeatingRobot/MicroBeatingRobot/Databank"
files = glob.glob(path + "/data_displacement_k*.csv")
files.sort()
df_list = []
for data in files:
    df = pd.read_csv(data)
    numid = re.findall("k=.*c" ,data)
    df = df.rename(columns={"displacement_x":"x_" + str(numid[0])[0:-2],"displacement_y":"y_"+ str(numid[0])[0:-2] })
    df_list.append(df)
df = pd.concat(df_list,axis = 1)
df = df.drop(['step', 'displacement_z'],axis = 1)
df = df.drop(index = 0)
for column_x, column_y in zip(df.filter(like = 'x_'), df.filter(like = 'y_')):
    df[column_x] -= 2
    df[column_y] += 2
    
```
