```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import re
import sys
%matplotlib tk
from mpl_toolkits.mplot3d import axes3d  # Axes3D のために必要
```

```python
path = "/home/tokoro/Desktop/MicroBeatingRobot/MicroBeatingRobot/Databank"
files_f = glob.glob(path + "/data_displacement_f*.csv")
files_k = glob.glob(path + "/data_displacement_k*.csv")

files_f.sort()
files_k.sort()

df_f_list = []
df_k_list = []

for data_f, data_k in zip(files_f, files_k):
    df_f = pd.read_csv(data_f)
    df_k = pd.read_csv(data_k) 
    id_f = re.findall("f=.*c" ,data_f)
    id_k = re.findall("k=.*c" ,data_k)
    
    df_f = df_f.rename(columns={"displacement_x":"x_" + str(id_f[0])[0:-2],"displacement_y":"y_"+ str(id_f[0])[0:-2] })
    df_k = df_k.rename(columns={"displacement_x":"x_" + str(id_k[0])[0:-2],"displacement_y":"y_"+ str(id_k[0])[0:-2] })
    
    df_f_list.append(df_f)
    df_k_list.append(df_k)

df_f = pd.concat(df_f_list,axis = 1)
df_k = pd.concat(df_k_list,axis = 1)

df = pd.concat([df_f,df_k],axis = 1)

df = df.drop(['step', 'displacement_z'],axis = 1)
df = df.drop(index = 0)
for column_x, column_y in zip(df.filter(like = 'x_'), df.filter(like = 'y_')):
    df[column_x] -= 2
    df[column_y] += 2
    
df
```

```python
df_cp = df.copy()
df.loc['mean'] = df.mean()
df_x = df.filter(like = 'x_f=')
df_y = df.filter(like = 'y_f=')
df_x = df_x.drop(columns = ['x_f=0.01','x_f=0.012'])
df_y = df_y.drop(columns = ['y_f=0.01','y_f=0.012'])

column_list_fx = df_x.columns
column_list_fy = df_y.columns

displacement_f = []

for xmean, ymean in zip(column_list_fx,column_list_fy):
    dis_mean = np.sqrt(df[xmean]['mean']**2 + df[ymean]['mean']**2)
    
    displacement_f.append(dis_mean)
```

```python
df.loc["displacement_per_piriod"] = 0
```

```python
for column in df.columns:
    key = 1 / float(column[4:])
    if "f=" in column:
        df[column]["displacement_per_piriod"] = df[column]["mean"] / key
    elif "k=" in column:
        df[column]["displacement_per_piriod"] = df[column]["mean"] / 50.0
    else:
        print('ERROR NOTHING COLUMN')
```

```python
df_x = df.filter(like = 'x_k=')
df_y = df.filter(like = 'y_k=')

column_list_kx = df_x.columns
column_list_ky = df_y.columns

displacement_k = []

for xmean, ymean in zip(column_list_kx,column_list_ky):
    dis_mean = np.sqrt(df[xmean]['mean']**2 + df[ymean]['mean']**2)
    
    displacement_k.append(dis_mean)
    
```
