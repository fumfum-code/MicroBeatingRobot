```python
path = "/home/tokoro/Desktop/MicroBeatingRobot/MicroBeatingRobot/Databank"
files_f = glob.glob(path + "/data_displacement_f*.csv")
files_k = glob.glob(path + "/data_displacement_k*.csv")

files_f.sort()
files_k.sort()

df_f_list = []
df_k_list = []

for data_f, data_k in zip(files_f, files_k):
    df_f = pd.read_csv(data_f)
    df_k = pd.read_csv(data_k) 
    id_f = re.findall("f=.*c" ,data_f)
    id_k = re.findall("k=.*c" ,data_k)
    
    df_f = df_f.rename(columns={"displacement_x":"x_" + str(id_f[0])[0:-2],"displacement_y":"y_"+ str(id_f[0])[0:-2] })
    df_k = df_k.rename(columns={"displacement_x":"x_" + str(id_k[0])[0:-2],"displacement_y":"y_"+ str(id_k[0])[0:-2] })
    
    df_f_list.append(df_f)
    df_k_list.append(df_k)

df_f = pd.concat(df_f_list,axis = 1)
df_k = pd.concat(df_k_list,axis = 1)

df = pd.concat([df_f,df_k],axis = 1)

df = df.drop(['step', 'displacement_z'],axis = 1)
df = df.drop(index = 0)
for column_x, column_y in zip(df.filter(like = 'x_'), df.filter(like = 'y_')):
    df[column_x] -= 2
    df[column_y] += 2
    
df
```

```python
df_cp = df.copy()
df.loc['mean'] = df.mean()
df_x = df.filter(like = 'x_f=')
df_y = df.filter(like = 'y_f=')
#df_x = df_x.drop(columns = ['x_f=0.01','x_f=0.012'])
#df_y = df_y.drop(columns = ['y_f=0.01','y_f=0.012'])

column_list_fx = df_x.columns
column_list_fy = df_y.columns

displacement_f = []

for xmean, ymean in zip(column_list_fx,column_list_fy):
    dis_mean = np.sqrt(df[xmean]['mean']**2 + df[ymean]['mean']**2)
    
    displacement_f.append(dis_mean)
```

```python
column_list_f = [float(x[4:]) for x in column_list_fx]
column_list_f
```

```python
dic = []
for i ,displacement in zip(column_list_f,displacement_f):
    a = dict(frec = i,K = 10.0, mean_displacement = displacement)
    dic.append(a)
```

```python
df_x = df.filter(like = 'x_k=')
df_y = df.filter(like = 'y_k=')

column_list_kx = df_x.columns
column_list_ky = df_y.columns

displacement_k = []

for xmean, ymean in zip(column_list_kx,column_list_ky):
    dis_mean = np.sqrt(df[xmean]['mean']**2 + df[ymean]['mean']**2)
    
    displacement_k.append(dis_mean)

column_list_k = [float(k[4:]) for k in column_list_kx]

for i ,displacement in zip(column_list_k,displacement_k):
    a = dict(frec = 0.02,K = i, mean_displacement = displacement)
    dic.append(a)
```

```python
files_fk = glob.glob(path + "/data_displacement_K*.csv")
files_fk.sort()
df_fk_list = []
for data_fk in files_fk:
    df_fk = pd.read_csv(data_fk)
    
    id_f = re.findall("f=.*c" ,data_fk)
    id_k = re.findall("K=.*c" ,data_fk)
    df_fk = df_fk.rename(columns={"displacement_x":"x_" + str(id_f[0])[0:-2] +"_"+str(id_k[0])[0:5],"displacement_y":"y_"+ str(id_f[0])[0:-2]+"_"+str(id_k[0])[0:-2]})
    
    df_fk_list.append(df_fk)

df_fk = pd.concat(df_fk_list,axis = 1)
df_fk = df_fk.drop(['step', 'displacement_z'],axis = 1)
df_fk = df_fk.drop(index = 0)
for column_x, column_y in zip(df_fk.filter(like = 'x_'), df_fk.filter(like = 'y_')):
    df_fk[column_x] -= 2
    df_fk[column_y] += 2
df_fk_cp = df_fk.copy()
df_fk.loc['mean'] = df_fk.mean()
df_x_fk = df_fk.filter(like = 'x_')
df_y_fk = df_fk.filter(like = 'y_')
#df_fk.rename(columns = {"x_*": } ,inplace = True)

df_x_fk.drop(range(1,20001))

f = []
k = []

for column_name in df_x_fk.columns:
    m = re.findall(r"\d.\d+",column_name)
    f.append(m[0])
    k.append(m[1])


disp = []
for xmean, ymean in zip (df_x_fk.columns,df_y_fk.columns):
    dis = np.sqrt(df_fk[xmean]["mean"]**2 + df_fk[ymean]['mean']**2)
    disp.append(dis)
i = 0

df_fk_total = pd.DataFrame(data = 0.0, columns = df_x_fk.columns, index = df_x_fk.columns)
for index_f , column_k in zip(df_x_fk.columns,df_x_fk.columns):
    df_fk_total[column_k][index_f] = disp[i]
    i+=1
    
f = [float(i) for i in f]
k = [float(i) for i in k]
df_fk_total.columns = k
df_fk_total.index = f
df_fk
```

```python
for kk ,F ,dis in zip(k,f,disp):
    a = dict(frec = F, K =kk ,mean_displacement = dis)
    dic.append(a)
for k in range(10):
    a = dict(frec = 0.01,)
```

```python
df_diagram = pd.DataFrame(data = dic)
df_diagram["flag"] = df_diagram['K']*0 + 1
```

```python
df_diagram.loc[df_diagram['K'] == 1.0, 'flag'] = 2
df_diagram.loc[df_diagram['K'] == 2.0, 'flag'] = 3
df_diagram.loc[df_diagram['frec'] < 0.015, 'flag'] = 4
df_diagram.loc[df_diagram['K'] < 1.0, 'flag'] = 4
```

```python
df_diagram
```

```python
df1 = df_diagram[df_diagram['flag'] == 1]
df2 = df_diagram[df_diagram['flag'] == 2]
df3 = df_diagram[df_diagram['flag'] == 3]
df4 = df_diagram[df_diagram['flag'] == 4]
```

```python
df_graph = df_diagram.pivot_table(values = "flag", index = ["frec"], columns = ["K"])
df_graph = df_graph.fillna(0)
df_graph
```

```python
df_origin = df_graph.reset_index().melt(id_vars=['frec'])
df_origin[df_origin['value'] == 0.0] = 1
df_origin.loc[df_origin['K'] == 1.0, 'value'] = 2
df_origin.loc[df_origin['K'] == 2.0, 'value'] = 3
df_origin.loc[df_origin['frec'] < 0.015, 'value'] = 4
df_origin.loc[df_origin['K'] < 1.0, 'value'] = 4

df1 = df_origin[df_origin['value'] == 1]
df2 = df_origin[df_origin['value'] == 2]
df3 = df_origin[df_origin['value'] == 3]
df4 = df_origin[df_origin['value'] == 4]

df1['K']
```

```python
fig, ax = plt.subplots()

ax.set_title('phase diagram')
ax.set_xlabel('K*')
ax.set_ylabel('frecency*')

ax.set_xlim([0,12])
ax.set_ylim([0,0.25])
ax.legend()

c1 = "propel"
c2 = "rotarion"
c3 = "overlap"
c4 = "stay"

ax.scatter(df1['K'], df1['frec'], color = "red", label = c1)
ax.scatter(df2['K'], df2['frec'], color = "blue", label = c2)
ax.scatter(df3['K'], df3['frec'], color = "green", label = c3)
ax.scatter(df4['K'], df4['frec'], color = "yellow", label = c4)


ax.legend()
#ax.contourf(df_graph)
```
