```python
import numpy as np
from numpy import linalg as LA
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import re
import sys
%matplotlib tk
from mpl_toolkits.mplot3d import axes3d  # Axes3D のために必要
```

```python
path = "/home/tokoro/Desktop/MicroBeatingRobot/MicroBeatingRobot/Databank"
files_f = glob.glob(path + "/data_displacement_f*.csv")
files_k = glob.glob(path + "/data_displacement_k*.csv")

files_f.sort()
files_k.sort()

df_f_list = []
df_k_list = []

for data_f, data_k in zip(files_f, files_k):
    df_f = pd.read_csv(data_f)
    df_k = pd.read_csv(data_k) 
    id_f = re.findall("f=.*c" ,data_f)
    id_k = re.findall("k=.*c" ,data_k)
    
    df_f = df_f.rename(columns={"displacement_x":"x_" + str(id_f[0])[0:-2],"displacement_y":"y_"+ str(id_f[0])[0:-2] })
    df_k = df_k.rename(columns={"displacement_x":"x_" + str(id_k[0])[0:-2],"displacement_y":"y_"+ str(id_k[0])[0:-2] })
    
    df_f_list.append(df_f)
    df_k_list.append(df_k)

df_f = pd.concat(df_f_list,axis = 1)
df_k = pd.concat(df_k_list,axis = 1)

df = pd.concat([df_f,df_k],axis = 1)

df = df.drop(['step', 'displacement_z'],axis = 1)
df = df.drop(index = 0)
for column_x, column_y in zip(df.filter(like = 'x_'), df.filter(like = 'y_')):
    df[column_x] -= 2
    df[column_y] += 2
    
df
```

```python
files_fk = glob.glob(path + "/data_displacement_K*.csv")
files_fk.sort()
df_fk_list = []
for data_fk in files_fk:
    df_fk = pd.read_csv(data_fk)
    
    id_f = re.findall("f=.*c" ,data_fk)
    id_k = re.findall("K=.*c" ,data_fk)
    df_fk = df_fk.rename(columns={"displacement_x":"x_" + str(id_f[0])[0:-2] +"_"+str(id_k[0])[0:5],"displacement_y":"y_"+ str(id_f[0])[0:-2]+"_"+str(id_k[0])[0:-2]})
    
    df_fk_list.append(df_fk)

df_fk = pd.concat(df_fk_list,axis = 1)
df_fk = df_fk.drop(['step', 'displacement_z'],axis = 1)
df_fk = df_fk.drop(index = 0)
for column_x, column_y in zip(df_fk.filter(like = 'x_'), df_fk.filter(like = 'y_')):
    df_fk[column_x] -= 2
    df_fk[column_y] += 2
df_fk_cp = df_fk.copy()
df_fk.loc['mean'] = df_fk.mean()
df_x_fk = df_fk.filter(like = 'x_')
df_y_fk = df_fk.filter(like = 'y_')
#df_fk.rename(columns = {"x_*": } ,inplace = True)
df_x_fk
```
