```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import re
%matplotlib tk
```

```python
path = "/home/tokoro/Desktop/MicroBeatingRobot/MicroBeatingRobot/Databank"
files = glob.glob(path + "/data_displacement_f*.csv")
files.sort()
df_list = []
for data in files:
    df = pd.read_csv(data)
    numid = re.findall("f=.*c" ,data)
    df = df.rename(columns={"displacement_x":"x_" + str(numid[0])[0:-2],"displacement_y":"y_"+ str(numid[0])[0:-2] })
    df_list.append(df)
df = pd.concat(df_list,axis = 1)
df = df.drop(['step', 'displacement_z'],axis = 1)
df = df.drop(index = 0)
for column_x, column_y in zip(df.filter(like = 'x_'), df.filter(like = 'y_')):
    df[column_x] -= 2
    df[column_y] += 2
df
```

```python
df.describe()
```

```python
df_cp = df.copy()
df.loc['mean'] = df.mean()
df_x = df.filter(like = 'x_f=')
df_y = df.filter(like = 'y_f=')
df_x = df_x.drop(columns = ['x_f=0.01','x_f=0.012'])
df_y = df_y.drop(columns = ['y_f=0.01','y_f=0.012'])
```

```python
column_list_x = df_x.columns
column_list_y = df_y.columns
```

```python
displacement = []

for xmean, ymean in zip(column_list_x,column_list_y):
    dis_mean = np.sqrt(df[xmean]['mean']**2 + df[ymean]['mean']**2)
    
    displacement.append(dis_mean)

print(displacement)
```

```python
column_list_x
```

```python
f = []

for fx in column_list_x:
    value = fx[4:]
    f.append(float(value))

print(f)
```

```python
df_mean = pd.DataFrame(f, columns = ['frec'])
df_mean['mean_displacement'] = displacement
df_mean = df_mean.sort_values('frec')
df_mean.set_index('frec', inplace = True)
df_mean
```

```python
fig, ax = plt.subplots()

ax.plot(df_mean,color = "#1f77b4",zorder = 1)
ax.scatter(df_mean.index ,df_mean["mean_displacement"],color = "#ff7f0e", marker = "o", zorder = 2)
ax.set_xlabel('frec')
ax.set_ylabel('mean_displacement')
ax.grid()
plt.savefig('frec_displacement.png')
plt.show()
```

```python
fig, ax = plt.subplots()
for x, y in zip (column_list_x,column_list_y):
    
    numid = float(x[4:])
    ax.plot(df_cp[x],df_cp[y], label = "f={}".format(numid))
    ax.set_xlabel('x_displacement')
    ax.set_ylabel('y_displacement')
    ax.set_xlim([-4,18])
    ax.set_ylim([-3,3])
    
    
    ax.legend()
    ax.set_title("MicroRobot trejectory")
    #plt.savefig('frec_all.png')
    plt.show()
    
```
